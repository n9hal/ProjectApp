```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('dataset_modified.csv')
```

```python
dataset.head()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding,LSTM,Dense, Bidirectional, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot 
```

```python
dataset = dataset.dropna()
```

```python
dataset.head()
```

```python
x= dataset.drop('label',axis=1)
```

```python
x.head()
```

```python
label = dataset['label']
```

```python
label.head()
```

```python
label.value_counts()
```

```python
x.shape
```

```python
label.shape
```

```python
voc_size=5000
```

```python
msg=x.copy()
msg['title'][5]
```

```python
msg.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range (0, len(msg)):
    review= re.sub('[^a-zA-z]',' ',msg['title'][i])
    review= review.lower()
    review=review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
print(i)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
##model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),label.shape
```

```python
x_final=np.array(embedded_docs)
label_final=np.array(label)
```

```python
x_final.shape,label_final.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final, label_final, test_size=0.2, random_state=27)
```

```python
#test model
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=5,batch_size=64)
```

```python
#Performance Matrix and Accuracy
y_pred=(model.predict(x_test)>0.5).astype("int32")
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```
